1. drop duplicates

```python
df = df.drop_duplicates()
### menghapus data duplikat
```

```python
df['last_review']
```

```python
df['last_review'] = pd.to_datetime(df['last_review'], format = '%m/%d/%Y')
```

mencari data yang null / NaN

```python
df[df['host_name'].isna()]
```

```python
df = df[~df['host_name'].isna()]
df.info()
# ~ artinya NOT, code diatas untuk memilih semua data dimana host_name tidak null kemudian menyimpan kedalam variabel df
```

```python
df.boxplot('price')
## cek data outliers
```

```python
df['price'].describe()
```

rumus menghapus outliers :
batas atas = Q3 + 1.5 * (Q3-Q1)
batas bawah = Q1 - 1.5 * (Q3-Q1)

```python
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
upper_limit = q3 + 1.5 * (q3-q1)
lower_limit = q1 - 1.5 * (q3-q1)
```

```python
df = df[df['price'] > lower_limit]
df = df[df['price'] < upper_limit]
df['price'].describe()
```

```python
df.boxplot('price')
```

```python
df.info()
```

```python
df['latitude'] = df['latitude'].astype('str')
df['longitude'] = df['longitude'].astype('str')
```

```python
df['Lat, Long'] = df['latitude']+","+df['longitude']
df
```
